```python
%matplotlib widget
import matplotlib.pyplot as plt
import nest_asyncio

nest_asyncio.apply()  # used in notebook

from phyling.ble import NanoPhyling, PhylingDevices
```

```python
# Define all devices
# gyro_offsets can be set after running devices.calibrate_gyro() below
devices = PhylingDevices(
    [
        NanoPhyling(
            name="NanoPhyling_18",
            config={"rate": 200, "data": ["acc_x", "acc_y", "acc_z", "gyro_x", "gyro_y", "gyro_z"]},
            gyro_offsets={"x": 0, "y": 0, "z": 0},
            display_name=None,  # optional display name for plotting (defaults nano-18, nano-19, etc. based on name if None)
        ),
        NanoPhyling(
            name="NanoPhyling_19",
            config={"rate": 200, "data": ["acc_x", "acc_y", "acc_z", "gyro_x", "gyro_y", "gyro_z"]},
            gyro_offsets={"x": 0, "y": 0, "z": 0},
            display_name=None,
        ),
        NanoPhyling(
            name="NanoPhyling_20",
            config={"rate": 200, "data": ["acc_x", "acc_y", "acc_z", "gyro_x", "gyro_y", "gyro_z"]},
            gyro_offsets={"x": 0, "y": 0, "z": 0},
            display_name=None,
        ),
        NanoPhyling(
            name="NanoPhyling_21",
            config={"rate": 200, "data": ["acc_x", "acc_y", "acc_z", "gyro_x", "gyro_y", "gyro_z"]},
            gyro_offsets={"x": 0, "y": 0, "z": 0},
            display_name=None,
        ),
    ],
    output_fs=200,  # resampling frequency for the fused DataFrame
)
```

```python
# Calibrate all devices simultaneously (~5 seconds, do not move any device)
devices.calibrate_gyro()

# Update gyro_offsets in the cells above with the printed values, then re-instantiate
# to persist calibration across sessions.
```

```python
# Record all devices simultaneously with a common T=0
devices.run(duration=30)
```

```python
# Fused DataFrame: columns T, time, nano-42.acc_x, ..., nano-43.acc_x, ...
df = devices.get_df(drop_nan=False)
df.head()
```

```python
# Plot acc_z for all devices on the same time axis
plt.figure()
for device in devices.devices:
    data_name = f"{device.get_name()}.acc_z"
    plt.plot(df["T"], df[data_name], label=data_name)

plt.xlabel("Time (s)")
plt.ylabel("Acc. Z (m/s²)")
plt.legend()
plt.grid()
plt.tight_layout()
```
