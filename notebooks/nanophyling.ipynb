```python
%matplotlib widget
import matplotlib.pyplot as plt

import nest_asyncio
nest_asyncio.apply()  # used in notebook

from phylib.nanophyling import NanoPhyling
```

```python
# You can get addresses when the sensor is connected (in run function)
# If you send the address instead of the name, you can skip the scan BLE step
# addresses = {
#     "NanoPhyling_42": "A53B406F-BBA2-538C-617D-861FC5DC6A7E",
# }
nano = NanoPhyling(
    name="NanoPhyling_42",
    # address=addresses["NanoPhyling_42"],
    config={
        "rate": 200,
        "data": ["acc_x", "acc_y", "acc_z", "gyro_x", "gyro_y", "gyro_z"],
    }
)
```

```python
# if duration is None, it will run until you stop it
nano.run(duration=10)
```

```python
df = nano.get_df()
```

```python
df
```

```python
plt.figure()
plt.plot(df["T"], df["acc_x"], label="acc_x")
plt.plot(df["T"], df["acc_y"], label="acc_y")
plt.plot(df["T"], df["acc_z"], label="acc_z")
```

```python
plt.figure()
plt.plot(df["T"], df["gyro_x"], label="gyro_x")
plt.plot(df["T"], df["gyro_y"], label="gyro_y")
plt.plot(df["T"], df["gyro_z"], label="gyro_z")
```
